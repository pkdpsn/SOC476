```python
import pandas as pd
import folium

# Sample data
data = {
    'ZIP': [10001, 20001, 30001],
    'Latitude': [40.7421, 38.8951, 33.7490],
    'Longitude': [-73.9897, -77.0364, -84.3880],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a Folium map centered at the mean latitude and longitude
m = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=5)

# Add markers for each ZIP code
for idx, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=f"ZIP: {row['ZIP']}").add_to(m)

# Display the map
m
```

```python
import pandas as pd
import numpy as np
import folium
from folium.plugins import HeatMap

# Sample data: random values for each state
np.random.seed(0)
states = ["New York", "California", "Texas", "Florida", "Illinois"]
values = np.random.rand(len(states))

# Create a DataFrame
data = pd.DataFrame({"State": states, "Value": values})
latitudes = [40.7128, 36.7783, 31.9686, 27.9944, 40.6331]
longitudes = [-74.0060, -119.4179, -99.9018, -81.7603, -89.3985]
data["Latitude"] = latitudes
data["Longitude"] = longitudes
# Create a Folium map centered at the mean latitude and longitude of the USA
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Define a function to get state boundaries
def get_state_boundaries(state):
    return folium.GeoJson(
        f"https://raw.githubusercontent.com/python-visualization/folium/master/examples/data/us-states.json",
        style_function=lambda feature: {
            "fillColor": "white",
            "color": "black",
            "weight": 2,
            "fillOpacity": 0.3 if feature["properties"]["name"] == state else 0,
        },
    )

# Add state boundaries to the map
for state in states:
    get_state_boundaries(state).add_to(m)

# Create a HeatMap layer based on the random values for each state
heat_data = [[row["Latitude"], row["Longitude"], row["Value"]] for idx, row in data.iterrows()]
HeatMap(heat_data, radius=20).add_to(m)

# Display the map
m
```

```python

import pandas as pd
import folium
from folium.plugins import HeatMap

# Load the data
data = pd.read_csv("us_income_zipcode.csv")

# Extract relevant columns
zipcode = data["ZIP"]
households_less_than_10000 = data["Households Less Than $10,000"]

# Normalize data (optional)
# households_normalized = (households_less_than_10000 - households_less_than_10000.min()) / (households_less_than_10000.max() - households_less_than_10000.min())

# Create a map centered on the USA
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Combine data into a list of [latitude, longitude, weight] for HeatMap
heat_data = list(zip(zipcode, households_less_than_10000))

# Plot heatmap
HeatMap(heat_data, radius=15).add_to(m)

# Display the map
m
```

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
from uszipcode import SearchEngine

# Load the data
data = pd.read_csv("us_income_zipcode.csv")

# Initialize the ZIP code search engine
search = SearchEngine()

# Convert ZIP codes to latitude and longitude
data["Latitude"] = data["ZIP"].apply(lambda x: search.by_zipcode(x).lat)
data["Longitude"] = data["ZIP"].apply(lambda x: search.by_zipcode(x).lng)

# Create a map centered on the USA
m = folium.Map(location=[37.0902, -95.7129], zoom_start=4)

# Combine data into a list of [latitude, longitude, weight] for HeatMap
heat_data = [[row["Latitude"], row["Longitude"], row["Households Less Than $10,000"]] for idx, row in data.iterrows()]

# Plot heatmap
HeatMap(heat_data, radius=15).add_to(m)

# Display the map
m
```

```python
import pandas as pd
import plotly.express as px
from uszipcode import SearchEngine

# Load the data
data = pd.read_csv("us_income_zipcode.csv")

# Initialize the ZIP code search engine
search = SearchEngine()

# Convert ZIP codes to state codes
data["State"] = data["ZIP"].apply(lambda x: search.by_zipcode(str(x)).state)
print(data["State"])

# Calculate the total income for each state
state_income = data.groupby("State")["Households Less Than $10,000"].sum().reset_index()

# Create a choropleth map
fig = px.choropleth(state_income, 
                    locations='State', 
                    color="Households Less Than $10,000",
                    locationmode = 'USA-states', 
                    scope="usa",
                    title='Households with income less than $10,000 by State',
                    color_continuous_scale=px.colors.sequential.Plasma)

fig.show()
```

```python
import pandas as pd
import plotly.express as px
from uszipcode import SearchEngine

# Load the data
data = pd.read_csv("us_income_zipcode.csv")

# Initialize the ZIP code search engine
search = SearchEngine()

# Convert ZIP codes to latitude, longitude, and state
data["Latitude"] = data["ZIP"].apply(lambda x: search.by_zipcode(str(x)).lat)
data["Longitude"] = data["ZIP"].apply(lambda x: search.by_zipcode(str(x)).lng)
data["State"] = data["ZIP"].apply(lambda x: search.by_zipcode(str(x)).state)

# Create a scatter plot on a map
fig = px.scatter_geo(data, 
                     lat='Latitude', 
                     lon='Longitude', 
                     color="Households Less Than $10,000",
                     hover_name='ZIP', 
                     scope="usa",
                     title='Households with income less than $10,000 by ZIP code',
                     color_continuous_scale=px.colors.sequential.Plasma)

fig.show()
```
