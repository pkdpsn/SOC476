```python

def make_map(feature, year, title):
    # Filter the data for the specified year
    year_data = data[data['Year'] == year]
    year_data[feature] *= year_data['Households']
    
    # Calculate the total for the feature for each state
    state_data = year_data.groupby("State")[feature].sum().reset_index()
    state_households = year_data.groupby("State")["Households"].sum().reset_index()

    # Calculate the proportion of the feature per household for each state
    state_data[feature] = state_data[feature] / state_households["Households"]
    a= float(state_data[feature].min())
    b = float(state_data[feature].max())
    print(a,b)
    # print(state_data)

    # Create a choropleth map
    fig = px.choropleth(state_data, 
                        locations='State',
                        title=title, 
                        color=feature,
                        locationmode='USA-states', 
                        scope="usa",
                        color_continuous_scale="Viridis",  # Use the Viridis color scale
                        range_color=(0,10) ) # Set the color scale range to min and max values of the feature

    # Remove color bar label, adjust title and margin, and set title font size
    fig.update_layout(
        coloraxis_colorbar=dict(title=''),
        title_x=0.5,  # Center the title
        margin=dict(l=0, r=0, t=30, b=0),  # Reduce margins to bring heatmap closer
        title_font=dict(size=18)  # Set title font size
    )

    # Save the figure to a file with a specified width and height
    filepath = os.path.join(os.getcwd(), "img", feature + "_" + str(year) + ".png")

    print(f"Saving map to {filepath}")
    fig.write_image(filepath, width=800, height=450)

    fig.show()
```

```python

def make_map_below_25000(year, title):
    # Filter the data for the specified year
    year_data = data[data['Year'] == year]
    year_data["Households Less Than $10,000"] *= year_data['Households']
    year_data["Households $10,000 to $14,999"] *= year_data['Households']
    year_data["Households $15,000 to $24,999"] *= year_data['Households']
    
    # Calculate the total households with income below $25,000 for each state
    year_data['Households < $25,000'] = year_data['Households Less Than $10,000'] + year_data['Households $10,000 to $14,999'] + year_data['Households $15,000 to $24,999']
    state_households_below_25000 = year_data.groupby("State")["Households < $25,000"].sum().reset_index()
    
    # Calculate the total households for each state
    state_households = year_data.groupby("State")["Households"].sum().reset_index()
    
    # Calculate the proportion of households below $25,000 for each state
    state_households_below_25000['Proportion'] = state_households_below_25000['Households < $25,000'] / state_households['Households'] 
    # print(state_households_below_25000['Proportion'])
    a= float(state_households_below_25000['Proportion'].min())
    b = float(state_households_below_25000['Proportion'].max())
    print(a,b)
    # Create a choropleth map
    fig = px.choropleth(state_households_below_25000, 
                        locations='State',
                        title=title, 
                        color='Proportion',
                        locationmode='USA-states', 
                        scope="usa",
                        color_continuous_scale="Viridis",  # Use the Viridis color scale
                        range_color=(10, 33))  # Set the color scale range from 0% to 100%

    # Remove color bar label, adjust title and margin, and set title font size
    fig.update_layout(
        coloraxis_colorbar=dict(title=''),
        title_x=0.5,  # Center the title
        margin=dict(l=0, r=0, t=30, b=0),  # Reduce margins to bring heatmap closer
        title_font=dict(size=18)  # Set title font size
    )

    # Save the figure to a file with a specified width and height
    filepath = os.path.join(os.getcwd(), "img", "Households_Below_25000_" + str(year) + ".png")
    print(f"Saving map to {filepath}")
    fig.write_image(filepath, width=800, height=450)

    fig.show()

```

```python
def make_map_mean_income_change(start_year, end_year, title):
    # Filter the data for the specified years
    start_year_data = data[data['Year'] == start_year]
    end_year_data = data[data['Year'] == end_year]
    
    # Calculate the total income for each state for the start year
    start_year_data['Total Income'] = start_year_data['Households'] * start_year_data['Households Mean Income (Dollars)']
    state_income_start = start_year_data.groupby("State")["Total Income"].sum().reset_index()
    
    # Calculate the total income for each state for the end year
    end_year_data['Total Income'] = end_year_data['Households'] * end_year_data['Households Mean Income (Dollars)']
    state_income_end = end_year_data.groupby("State")["Total Income"].sum().reset_index()
    
    # Calculate the mean income for each state for the start year
    state_income_start['Mean Income Start'] = state_income_start['Total Income'] / start_year_data.groupby("State")["Households"].sum().reset_index()['Households']
    
    # Calculate the mean income for each state for the end year
    state_income_end['Mean Income End'] = state_income_end['Total Income'] / end_year_data.groupby("State")["Households"].sum().reset_index()['Households']
    
    # Calculate the rate of change of mean income
    state_income_end['Rate of Change'] = (state_income_end['Mean Income End'] - state_income_start['Mean Income Start']) / state_income_start['Mean Income Start'] * 100

    # Create a choropleth map
    fig = px.choropleth(state_income_end, 
                        locations='State',
                        title=title, 
                        color='Rate of Change',
                        locationmode='USA-states', 
                        scope="usa",
                        color_continuous_scale="Viridis",  # Use the Viridis color scale
                        range_color=(0, 10))  # Set the color scale range from -20% to 20%

    # Remove color bar label, adjust title and margin, and set title font size
    fig.update_layout(
        coloraxis_colorbar=dict(title=''),
        title_x=0.5,  # Center the title
        margin=dict(l=0, r=0, t=30, b=0),  # Reduce margins to bring heatmap closer
        title_font=dict(size=18)  # Set title font size
    )

    # Save the figure to a file with a specified width and height
    filepath = os.path.join(os.getcwd(), "img", "Mean_Income_Change_" + str(start_year) + "_to_" + str(end_year) + ".png")
    print(f"Saving map to {filepath}")
    fig.write_image(filepath, width=800, height=450)

    fig.show()

# # Example usage:
# make_map_mean_income_change(2011, 2016, "Rate of Change of Mean Household Income from 2011 to 2016")
# make_map_mean_income_change(2016, 2021, "Rate of Change of Mean Household Income from 2016 to 2021")
```

```python
def make_map_rate_of_change(year1, year2, title):
    # Filter the data for the specified years
    year_data1 = data[data['Year'] == year1]
    year_data2 = data[data['Year'] == year2]
    
    # Calculate the total households with income below $25,000 for each state for both years
    year_data1["Households Below $25,000"] = year_data1["Households Less Than $10,000"] + \
                                              year_data1["Households $10,000 to $14,999"] + \
                                              year_data1["Households $15,000 to $24,999"]
    
    year_data2["Households Below $25,000"] = year_data2["Households Less Than $10,000"] + \
                                              year_data2["Households $10,000 to $14,999"] + \
                                              year_data2["Households $15,000 to $24,999"]
    
    state_households_below_25000_1 = year_data1.groupby("State")["Households Below $25,000"].sum().reset_index()
    state_households_below_25000_2 = year_data2.groupby("State")["Households Below $25,000"].sum().reset_index()
    
    # Merge the data for both years
    merged_data = pd.merge(state_households_below_25000_1, state_households_below_25000_2, on="State", suffixes=('_'+str(year1), '_'+str(year2)))
    
    # Calculate the rate of change
    merged_data['Rate of Change'] = ((merged_data['Households Below $25,000_'+str(year2)] - merged_data['Households Below $25,000_'+str(year1)]) / merged_data['Households Below $25,000_'+str(year1)]) * 100
    
    # Create a choropleth map
    fig = px.choropleth(merged_data, 
                        locations='State',
                        title=title, 
                        color='Rate of Change',
                        locationmode='USA-states', 
                        scope="usa",
                        color_continuous_scale="Viridis",  # Use the Viridis color scale
                        range_color=(-15, 5))  # Set the color scale range from -100% to 100%

    # Remove color bar label, adjust title and margin, and set title font size
    fig.update_layout(
        coloraxis_colorbar=dict(title=''),
        title_x=0.5,  # Center the title
        margin=dict(l=0, r=0, t=30, b=0),  # Reduce margins to bring heatmap closer
        title_font=dict(size=18)  # Set title font size
    )

    # Save the figure to a file with a specified width and height
    filepath = os.path.join(os.getcwd(), "img", "Rate_of_Change_" + str(year1) + "_to_" + str(year2) + ".png")
    print(f"Saving map to {filepath}")
    fig.write_image(filepath, width=800, height=450)

    fig.show()

# Call the function to plot the rate of change from 2011 to 2016
# make_map_rate_of_change(2011, 2016, "Rate of Change in Proportion of Households Below $25,000 from 2011 to 2016")

# # Call the function to plot the rate of change from 2016 to 2021

# make_map_rate_of_change(2016, 2021, "Rate of Change in Proportion of Households Below $25,000 from 2016 to 2021")
make_map_rate_of_change(2019, 2020, "Rate of Change in Proportion of Households Below $25,000 from 2019 to 2020")
make_map_rate_of_change(2020, 2021, "Rate of Change in Proportion of Households Below $25,000 from 2020 to 2021")

```
